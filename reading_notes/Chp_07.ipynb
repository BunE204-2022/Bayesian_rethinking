# 7.0 Prerequisites

- Ockham’s razor：越简单的模型是更好的模型。但在真实的实践中，奥卡姆剃刀原则的使用不太便利，因为当我们建立模型时总是会在**准确率(accuracy)** 与 **简洁性(simplicity)** 之间做出权衡，即  **bias-variance trade-off**。
- Overfitting：在训练集中过分拟合数据，但其在测试集中的准确率过低，即模型具有**high variance**，其泛化性能很弱。
- Underfitting：模型并没有很好地描述数据的性质，具有**low variance**，但其在训练集与测试集中都具有**high bias**。

<font color=Red>**本章即给出几种工具用于在bias和variance之间做出trade-off。**</font>

# 7.1 The problem with parameters

  - fit(模型拟合)：度量模型能过多好地预测数据。在高斯线性模型中，$R^2$是一个常用的指标：
  $$R^2=\frac{\operatorname{var}(\operatorname{outcome})-\operatorname{var}(\operatorname{residuals})}{\operatorname{var}(\operatorname{outcome})}=1-\frac{\operatorname{var}(\operatorname{residuals})}{\operatorname{var}(\text{outcome})}$$

  但这样的指标具有严重的弊端，因为当模型越复杂时，模型必然能够更好地拟合数据，$R^2$也会越接近1。

## 7.2.1 Firing the weatherperson

考虑以下场景：天气观察员通过观察今天的天气现象预测明天下雨的概率，以下是两名天气观察员连续十天的预测结果以及对应的观察结果。 

  

 以正确预测的平均概率为评估指标，第一位天气观察员gets 3 × 1 + 7 × 0.4 = 5.8 hits in 10 days, for a rate of 5.8/10 = 0.58 correct predictions per day. 第二位天气观察员gets 3 × 0 + 7 × 1 = 7, for 7/10 = 0.7 hits per day. The newcomer wins.

 这显然是不合理的，因为第二位天气观察员没有进行任何预测，只是认为每一天都不会下雨，但基于此种准则，Ta却表现得更加优秀。

### 7.2.1.1 Costs and benefits

从代价角度考虑另一种评估指标，做出以下假设。
- 假设你讨厌带伞而且讨厌在下雨天淋湿
- 假设你带伞的话幸福值会-1，在下雨天淋湿幸福值会-5.
- 假设你带伞的概率与预测下雨的概率相同。



第一位天气观察员得到-7.2的幸福值，第二位则是-15的幸福值。显然，第一位天气观察员更优秀。

### 7.2.1.2 Measuring accuracy

<font color=Red>**Question**：Which definition of “accuracy” is maximized by knowing the true model generating the data?</font>

- 作者指出考虑**联合概率(joint probability)**
  - 第一位观察员将整个事件预测正确的概率为：$1^3\times0.4^7\approx0.005$
  - 第二位观察员将整个事件预测正确的概率为：$0^3\times1^7= 0$
- **联合概率**即是我们想要的**Accuracy指标**。
  - Because it appears in Bayes’ theorem as the likelihood. It’s the unique measure that correctly counts up the relative numberof ways each event (sequence of rain and shine) could happen.
  - Another way to think of this is to consider what happens when we maximize average probability or joint probability. The true data-generating model will not have the highest hit rate. You saw this already with the weatherperson: Assigning zero probability to rain improves hit rate, but it is clearly wrong. In contrast, the true model will have the highest joint probability
  - 通常，我们对**联合概率**取对数处理。

## 7.2.2  Information and uncertainty && 7.2.3  From entropy to accuracy

<font color=Red>**Question**：How to measure distance from perfect prediction？</font>  
信息论指出，回答该问题可以先回答另一个问题：How much is our uncertainty reduced by learning an outcome?即当我们得知一个结果时，事件发生的不确定性减少了多少？

- Entropy熵：描述一个对数概率分布的不确定性
  $$H(p)=-\sum\limits_{i=1}^n p_i\log(p_i)=\mathbb{E}_{x\sim p}[-\log p]$$
  - Code 7.12假设某地下雨天与晴天的概率分别为0.3与0.7，其熵为0.61。
  - 若我们居住在沙特阿拉伯，此地下雨天与晴天的概率分别为0.01与0.99，其熵为0.06。
  - 熵的减小意味着不确定性的降低，这很好理解，因为沙特阿拉巴几乎不下雨，不确定性当然比较低。
- KL散度：衡量两个概率分布之间的距离
  $$D_{\text{KL}}(p,q)=\sum\limits_i p_i\big(\log(p_i)-\log(q_i)\big)=\sum\limits_i p_i\log\left(\frac{p_i}{q_i}\right)$$
  如Code 7.12图所示，KL散度所衡量的“距离”并非如欧式距离一般是对称的，其是非对称的，当且仅当$p=q$时，KL散度为0。
- Cross entropy交叉熵
  $$H(p,q)=-\sum\limits_{i=1}^n p_i\log(q_i)=\mathbb{E}_{x\sim p}[-\log q]$$

## 7.2.4   Estimating divergence

- 要获得KL散度，需要知道真实的联合概率p，但在事件未发生之前我们是无法获得p的，那么我们要如何计算距离$D_{KL}(p,q)$呢？
  - 我们的目标是获得一个最好的模型，假设我们从数据集中学的得到两个模型q与r，我们要做的其实是比较模型q和模型r孰优孰劣，即$D_{KL}(p,q)$与$D_{KL}(p,r)$谁更小。
  - 由KL散度的性质可推导出我们并不需要知道p也可以评价哪个模型更好，推导如下：
    $$D_{\text{KL}}(p,q)=\sum\limits_i p_i\big(\log(p_i)-\log(q_i)\big)=\sum\limits_i p_i\log\left(\frac{p_i}{q_i}\right)$$
    $$D_{\text{KL}}(p,r)=\sum\limits_i p_i\big(\log(p_i)-\log(r_i)\big)=\sum\limits_i p_i\log\left(\frac{p_i}{r_i}\right)$$
    $$D_{\text{KL}}(p,q)-D_{\text{KL}}(p,r)=\mathbb{E}_{x\sim p}[\log r]-\mathbb{E}_{x\sim p}[\log q]$$
  - 因此我们只需要估计$\mathbb{E}_{x\sim p}[\log r]$与$\mathbb{E}_{x\sim p}[\log q]$即可，可采用点估计方法
    $$S(q)=\sum\limits_i\log(q_i)$$
  - 在Bayes中我们需要使用整个后验分布，即$q_i$的后验分布，如Code 7.13和Code 7.14所示计算了模型m_7_1的$\log(q_i)$
    $$\operatorname{lppd}(y,\Theta)=\sum\limits_i\log\frac{1}{S}\sum\limits_s p(y_i|\Theta_s)$$
    $$\operatorname{DEVIANCE}=-2\operatorname{lppd}$$

## 7.2.5    Scoring the right data

- $S(q)=\sum\limits_i\log(q_i)$也有其弊端，如$R^2$一般，随着模型变得复杂，其值也会增大。接下来计算模型m_7_1~m_7_6的$S(q)$值。从Code 7.15可发现随着模型复杂度的增加，$S(q)$亦增大。

#### Code 7.15

```python
# make an lppd function that can be applied to all models (from code above)
def lppd(model, n_samples=1e4):
    n_samples = int(n_samples)

    pars = stats.multivariate_normal(model.params, model.cov_params()).rvs(n_samples).T
    dmat = dmatrix(
        model.model.data.design_info, brains, return_type="dataframe"
    ).values  # get model design matrix
    pred = dmat.dot(pars)

    n, ns = pred.shape

    # this approach for calculating lppd isfrom 7.14
    sigmas = (np.sum((pred - brains.brain_std.values.reshape(-1, 1)) ** 2, 0) / 7) ** 0.5
    ll = np.zeros((n, ns))
    for s in range(ns):
        logprob = stats.norm.logpdf(brains.brain_std, pred[:, s], sigmas[s])
        ll[:, s] = logprob

    lppd = np.zeros(n)
    for i in range(n):
        lppd[i] = logsumexp(ll[i]) - np.log(ns)

    return lppd
```

- Training Sample与Test Sample
  - Suppose there’s a training sample of size N.
  - Compute the posterior distribution of a model for the training sample, and compute the score on the training sample. Call this score $D_{train}$.
  - Suppose another sample of size N from the same process. This is the test sample.
  - Compute the score on the test sample, using the posterior trained on the training
sample. Call this new score $D_{test}$.
- 基于以下模型生成数据并训练模型，计算其在训练集与测试集中的deviance，N表示模拟次数
  
  \begin{array}{l}y_i\sim\operatorname{Normal}(\mu_i,1)\\ \mu_i=(0.15)x_{1,i}-(0.4)x_{2,i}\end{array} 
  - 左图：随着模型参数数量增加，训练集deviance下降，测试集在参数数量为3时deviance最小，生成数据的模型的参数即为3

-  While deviance on training data always improves with additional predictor variables, deviance on future data may or may not, depending upon both the true data-generating process and how much data is available to precisely estimate the parameters.

#### Code 7.16

This relies on the `sim.train.test` function in the `rethinking` package. [This](https://github.com/rmcelreath/rethinking/blob/master/R/sim_train_test.R) is the original function.

The python port of this function below is from [Rethinking/Chp_06](https://nbviewer.jupyter.org/github/pymc-devs/resources/blob/master/Rethinking/Chp_06.ipynb) Code 6.12.

```python
def sim_train_test(N=20, k=3, rho=[0.15, -0.4], b_sigma=100, samples=1000):

    n_dim = 1 + len(rho)
    if n_dim < k:
        n_dim = k
    Rho = np.diag(np.ones(n_dim))
    Rho[0, 1:3:1] = rho
    i_lower = np.tril_indices(n_dim, -1)
    Rho[i_lower] = Rho.T[i_lower]

    x_train = stats.multivariate_normal.rvs(cov=Rho, size=N)
    x_test = stats.multivariate_normal.rvs(cov=Rho, size=N)

    mm_train = np.ones((N, 1))

    np.concatenate([mm_train, x_train[:, 1:k]], axis=1)

    # Using pymc
    with pm.Model() as m_sim:
        vec_V = pm.MvNormal(
            "vec_V",
            mu=0,
            cov=b_sigma * np.eye(n_dim),
            shape=(1, n_dim),
            initval=np.random.randn(1, n_dim) * 0.01,
        )
        mu = pm.Deterministic("mu", 0 + pm.math.dot(x_train, vec_V.T))
        y = pm.Normal("y", mu=mu, sigma=1, observed=x_train[:, 0].reshape(-1, 1))

    with m_sim:
        trace_m_sim = pm.sample(samples, progressbar=False)

    vec = az.summary(trace_m_sim)["mean"][:n_dim]
    vec = np.array([i for i in vec]).reshape(n_dim, -1)

    dev_train = -2 * sum(stats.norm.logpdf(x_train, loc=np.matmul(x_train, vec), scale=1))

    mm_test = np.ones((N, 1))

    mm_test = np.concatenate([mm_test, x_test[:, 1 : k + 1]], axis=1)

    dev_test = -2 * sum(stats.norm.logpdf(x_test[:, 0], loc=np.matmul(mm_test, vec), scale=1))

    return np.mean(dev_train), np.mean(dev_test)
```

```python
# This cell is expected have a long run time
n = 20
tries = 10
param = 6
r = np.zeros(shape=(param - 1, 4))

train = []
test = []

# setting log output to error
logger = logging.getLogger("pymc")
init_level = logger.level
logger.setLevel(logging.ERROR)

for j in range(2, param + 1):
    print(f"Num Params: {j}")
    for i in tqdm.tqdm(range(1, tries + 1)):
        tr, te = sim_train_test(N=n, k=param, samples=1000)
        train.append(tr), test.append(te)
    r[j - 2, :] = (
        np.mean(train),
        np.std(train, ddof=1),
        np.mean(test),
        np.std(test, ddof=1),
    )

# resetting logger level
logger.setLevel(init_level)
```

# 7.3 Golem taming: regularization

- 贝叶斯与正则化的关系
  
  假设 $p(\theta) \sim \mathcal{N} (0, \eta^2 I)$,

  \begin{align*}
  \theta_{\mathrm{MAP}} & = \arg \max_\theta p(y \ \vert \ x, \theta) \ p(\theta) \\
                        & = \arg \min_\theta - \log p(y \ \vert \ x, \theta) - \log p(\theta) \\
                        & = \arg \min_\theta - \log p(y \ \vert \ x, \theta) - \log \frac{1}{(2 \pi)^{d / 2} \vert \Sigma \vert^{1/2}} \exp \big( -\frac{1}{2} (\theta - \mu)^T \Sigma^{-1} (\theta - \mu) \big) \\
                        & = \arg \min_\theta - \log p(y \ \vert \ x, \theta) + \frac{1}{2} \theta^T \Sigma^{-1} \theta \\
                        & = \arg \min_\theta - \log p(y \ \vert \ x, \theta) + \frac{1}{2 \eta^2} \Vert \theta \Vert_2^2 \\
                        & = \arg \min_\theta - \log p(y \ \vert \ x, \theta) + \lambda \Vert \theta \Vert_2^2
  \end{align*}

  其中 $\lambda = 1 / (2 \eta^2)$.
- When the priors are flat or nearly flat, the machine interprets this to mean that every parameter value is equally plausible. As a result, the model returns a posterior that encodes as much of the training sample—as represented by the likelihood function—as possible. 因此先验分布的选择十分重要，若先验分布太平，可能导致过拟合，若太尖，可能导致欠拟合。

#### Code 7.31

dSE is calculated by compare above, but `rethinking` produces a pairwise comparison. This is not implemented in `arviz`, but we can hack it together:

```python
dataset_dict = {"m_6_6": m_6_6_trace, "m_6_7": m_6_7_trace, "m_6_8": m_6_8_trace}

# compare all models
s0 = az.compare(dataset_dict, ic="waic", scale="deviance")["dse"]
# the output compares each model to the 'best' model - i.e. two models are compared to one.
# to complete a pair-wise comparison we need to compare the remaining two models.

# to do this, remove the 'best' model from the input data
del dataset_dict[s0.index[0]]

# re-run compare with the remaining two models
s1 = az.compare(dataset_dict, ic="waic", scale="deviance")["dse"]

# s0 compares two models to one model, and s1 compares the remaining two models to each other
# now we just nee to wrangle them together!

# convert them both to dataframes, setting the name to the 'best' model in each `compare` output.
# (i.e. the name is the model that others are compared to)
df_0 = s0.to_frame(name=s0.index[0])
df_1 = s1.to_frame(name=s1.index[0])

# merge these dataframes to create a pairwise comparison
pd.merge(df_0, df_1, left_index=True, right_index=True)
```

**Note:** this work for three models, but will get increasingly hack-y with additional models. The function below can be applied to *n* models:

```python
def pairwise_compare(dataset_dict, metric="dse", **kwargs):
    """
    Calculate pairwise comparison of models in dataset_dict.

    Parameters
    ----------
    dataset_dict : dict
        A dict containing two ore more {'name': pymc.backends.base.MultiTrace}
        items.
    metric : str
        The name of the matric to be calculated. Can be any valid column output
        by `arviz.compare`. Note that this may change depending on the **kwargs
        that are specified.
    kwargs
        Arguments passed to `arviz.compare`
    """
    data_dict = dataset_dict.copy()
    dicts = []

    while len(data_dict) > 1:
        c = az.compare(data_dict, **kwargs)[metric]
        dicts.append(c.to_frame(name=c.index[0]))
        del data_dict[c.index[0]]

    return pd.concat(dicts, axis=1)
```

```python
dataset_dict = {"m_6_6": m_6_6_trace, "m_6_7": m_6_7_trace, "m_6_8": m_6_8_trace}

pairwise_compare(dataset_dict, metric="dse", ic="waic", scale="deviance")
```
